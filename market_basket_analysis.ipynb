```python
import pandas as pd

print("Reading huge file... please wait...")

# 1. Read the big file in chunks to prevent crashing
chunk_size = 50000
germany_data = []

# Update 'online_retail_II.csv' if your file name is different
# encoding='ISO-8859-1' fixes special character errors
for chunk in pd.read_csv('online_retail_II.csv', encoding='ISO-8859-1', chunksize=chunk_size):
    # Filter: Keep only Germany
    filtered = chunk[chunk['Country'] == 'Germany']
    germany_data.append(filtered)

# 2. Combine and Save
final_df = pd.concat(germany_data)
final_df.to_csv('Germany_Data.csv', index=False)

print(f"Success! Created 'Germany_Data.csv' with {len(final_df)} rows.")
```
